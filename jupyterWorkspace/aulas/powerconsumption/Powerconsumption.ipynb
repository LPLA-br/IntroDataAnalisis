```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('powerconsumption.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
# Dispersão das frequências, Histograma, Frequência por unidade aproximadamente
sns.histplot(data=df,x="Temperature")
```

```python
# Adiciona duas colunas a partir de datetime -> mês, dia
df['mes'] = pd.to_datetime(df['Datetime']).dt.month
df['dia'] = pd.to_datetime(df['Datetime']).dt.dayofyear
```

```python
df.head()
```

```python
sns.lineplot(data=df, x='dia', y='Temperature')
```

```python
#gases de emissão geral
sns.lineplot(data=df, x='dia', y='GeneralDiffuseFlows')
```

```python
#gases de emissão local
sns.lineplot(data=df, x='dia', y='DiffuseFlows')
```
